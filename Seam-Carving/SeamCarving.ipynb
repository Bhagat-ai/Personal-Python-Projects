```python
import numpy as np
from PIL import Image
import seam_carving
import matplotlib.pyplot as plt

img = np.array(Image.open('BT.jpg'))
img_height, img_width, _ = img.shape
print("Original image size: "+ str(img_width)+"x"+str(img_height))
plt.imshow(img)
print("Please wait a few moments, your image is being processed....")
final = seam_carving.resize( img, (img_width - (img_width/2), img_height),
    energy_mode='backward',  
    order='width-first',  
    keep_mask=None
)
FINAL=Image.fromarray(final)
FINAL=FINAL.save("BT.jpg")
Image.fromarray(final).show()

print("Seam-Carved image size: "+ str(int(img_width-(img_width/2)))+"x"+str(img_height))
```
